```python
import pandas as pd 
import glob
import os
```

```python
# Path to your data folder
data_path = r'C:\Users\skman\Pictures\UK masters\Data Science\Projects\quantium-starter-repo\data'

# 1. Get a list of all CSV files in the folder
all_files = glob.glob(os.path.join(data_path, "*.csv"))

# 2. Read each file into a list of DataFrames
df_list = [pd.read_csv(file) for file in all_files]

# 3. Concatenate all DataFrames into one
df = pd.concat(df_list, ignore_index=True)

# Verify the result
print(f"Loaded {len(all_files)} files.")
df.head()
```

```python
df=df[df['product']=='pink morsel']
```

```python
# 1. Clean the price column (remove '$' and convert to float)
df['price'] = df['price'].replace('[$,]', '', regex=True).astype(float)

# 2. Create the sales column
df['sales'] = df['price'] * df['quantity']

# 3. Remove the original price and quantity columns
df.drop(columns=['price', 'quantity'], inplace=True)

# Display the result
df.head()
```
